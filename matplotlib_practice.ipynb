```python
%matplotlib inline
```

```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "Arp", "May", "June", "July", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total flare in US dollars accumulated for each month.Javascript
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label="Boston")
# Create labels for the x and y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45
plt.ylim(0,45)
# Create a title
plt.title("PyBer Fare by Month")
# Add a grid
plt.grid()
# Add a legend
plt.legend()
```

```python
# Skill Drill 5.1.4
plt.plot(x_axis, y_axis, marker="d", color="green", linewidth=2, label="Boston")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.ylim(0,45)
plt.title("PyBer Fare by Month")
plt.grid()
plt.legend()
```

```python
# Create the plot (Bar Graph)
plt.bar(x_axis, y_axis, color="green", label="Boston")
# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a Title
plt.title("PyBer Fare by Month")
#Add the legend.
plt.legend()
```

```python
# Horizontal bar chart
plt.barh(x_axis, y_axis)
```

```python
# Horizontal bar chart with x and y switched
plt.barh(y_axis, x_axis)
```

```python
# Plot the graph but have months in ascending order
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
plt.gca().invert_yaxis()
plt.xlabel("Fare($)")
plt.ylabel("Month")
plt.legend()
plt.title("PyBer Fare by Month")
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
plt.gca().invert_yaxis()
plt.xlabel("Fare($)")
plt.ylabel("Month")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
# Create Scatter plot using plt.plot
plt.plot(x_axis, y_axis, "o")
```

```python
# Create Scatter plot using plt.scatter
plt.scatter(y_axis, x_axis, color="red", label="Chicago")
plt.gca().invert_yaxis()
plt.xlim(0,50)
plt.title("PyBer Fare by Month")
plt.xlabel("Month")
plt.ylabel("Fare($)")
plt.legend()
```

```python
# Create a bubble graph
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Create bubble chart
plt.scatter(x_axis, y_axis, s=[i *3 for i in y_axis])
```

```python
# Create scatter plot using fig, ax
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create bubble chart using fig, ax
fig, ax = plt.subplots()
#Marker size 5 times eachdata point, black edge, color skyblue, 20% transparent, linewidth 2 points
ax.scatter(y_axis, x_axis, s=[i*5 for i in y_axis], color="skyblue", alpha=.2, edgecolors="black", linewidths=2, label="Boston")
plt.gca().invert_yaxis()
plt.legend()
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Month")
```

```python
# Create a pie chart
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Make the chart bigger
plt.subplots(figsize=(8,8))
# Add percentages for each month and "explode the largest percentage"
explode_values= (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
# Assign colors, one for each month
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, autopct="%.1f%%")
plt.show()
```

```python
# Create a pie chart using the object-Oriented interface
explode_values= (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis,labels=x_axis, colors=colors, explode= explode_values, shadow="True", startangle=90, autopct = "%.1f%%")
plt.gca().invert_xaxis()
plt.show()
```
