```python
%matplotlib inline
```

```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "Arp", "May", "June", "July", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total flare in US dollars accumulated for each month.Javascript
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label="Boston")
# Create labels for the x and y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45
plt.ylim(0,45)
# Create a title
plt.title("PyBer Fare by Month")
# Add a grid
plt.grid()
# Add a legend
plt.legend()
```

```python
# Skill Drill 5.1.4
plt.plot(x_axis, y_axis, marker="d", color="green", linewidth=2, label="Boston")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.ylim(0,45)
plt.title("PyBer Fare by Month")
plt.grid()
plt.legend()
```

```python
# Create the plot (Bar Graph)
plt.bar(x_axis, y_axis, color="green", label="Boston")
# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a Title
plt.title("PyBer Fare by Month")
#Add the legend.
plt.legend()
```

```python
# Horizontal bar chart
plt.barh(x_axis, y_axis)
```

```python
# Horizontal bar chart with x and y switched
plt.barh(y_axis, x_axis)
```

```python
# Plot the graph but have months in ascending order
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
plt.gca().invert_yaxis()
plt.xlabel("Fare($)")
plt.ylabel("Month")
plt.legend()
plt.title("PyBer Fare by Month")
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
plt.gca().invert_yaxis()
plt.xlabel("Fare($)")
plt.ylabel("Month")
plt.title("PyBer Fare by Month")
plt.legend()
```
